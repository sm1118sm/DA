```python
# Step 1 필요 모듈 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="seaborn")

# Step 2. 파일 불러오기
# 문제 1. 주어진 fish.csv 파일을 인코딩 문제를 해결하여 불러온 후 data1 변수에 할당하세요.
from google.colab import files
uploaded = files.upload()

# 파일 읽기
file_name = list(uploaded.keys())[0]  # 업로드한 파일 이름 가져오기

data1 = pd.read_csv(file_name)
display( data1.head( ) )
print(data1.shape)
```

```python
# Step 3. Species 값이 Bream 인 것만 추출하기
# 문제 2. 위 data1 변수 값에서 Species 값이 'Bream' 인 데이터만 추출한 후
# bream_data 변수에 할당하세요.
bream_data = data1[data1['Species'].isin(['Bream'])]
bream_data
```

```python
# Step 4. Bream 의 Weight 컬럼과 Height 컬럼 값만 추출하기
bream_Weight = bream_data['Weight']
bream_Height = bream_data['Height']

# Step 5. 산점도로 시각화하기
# 문제 3. 위 Step 4의 bream_Weight 변수와 bream_Height 변수 값을 사용하여
# 예시화면과 같이 산점도(산포도)그래프를 그리세요.
# 이때 matplotlib의 함수를 사용하세요.

plt.scatter(bream_Weight, bream_Height)

plt.title('Bream Length vs Width')
plt.xlabel('bream_Weight')
plt.ylabel('bream_Height')
plt.show()
```

```python
# 이번에는 위 Step 2 에서 생성한 data1 변수에서
# Species 값이 'Bream','Perch','Parkki' 인 데이터의 Weight 값과 Height 값을 추출한 후
# 예시 이미지와 같이 산점도 그래프로 시각화를 진행합니다

# Step 6. 추출할 종류 지정하고 색상 지정하기
species_list = ['Bream','Perch','Parkki']
print(species_list)
```

```python
# Step 7. 각 종류별로 색상과 무게와 높이 값 할당하여 그래프 그리기
# 종류별 색상 지정하기
colors = {'Bream': 'red','Perch' : 'green' ,'Parkki':'blue' }

for species in species_list:
    species_data = data1[data1['Species'] == species]
    # 문제 4. 윗줄에서 생성한 species_data의 'Weight' 값과 Height' 값을 사용하여
    # 예시와 같이 3가지 종류의 생선의 산점도(산포도)를 그리세요(색상은 위 셀의 colors 값 사용하세요)

    plt.scatter(species_data['Weight'], species_data['Height'], color=colors[species], label=species)

# Step 8. 부가 정보 출력하기
plt.title('Fish Weight vs Height by Species')
plt.xlabel('Fish Weight')
plt.ylabel('Fish Height')
plt.legend()
plt.show()
```

```python
# 위 문제 1에서 불러온 data1 변수에서
# Species 값이 'Bream','Perch','Roach','Parkki' 인 데이터의 Weight 값과 Height 값을 추출한 후
# 화면을 4등분하여 예시 이미지와 같이 산점도 그래프로 시각화 하세요.
# 색상은 자유롭게 지정하세요

#Step 9. Bream , Perch, Roach , Parkki 의 Weight 와 Height 값을 추출합니다.
fish1 = data1.loc[data1['Species'] == 'Bream', ['Weight','Height'] ]
#문제 5. Perch , Roach , Parkki 의 Weight , Height 값을 추출하여 각 변수에 저장하세요.
# 아래의 3곳에 들어갈 내용은 위 7번 줄의 내용을 참고하세요
fish2 = data1.loc[data1['Species'] == 'Perch', ['Weight','Height'] ]
fish3 = data1.loc[data1['Species'] == 'Roach', ['Weight','Height'] ]
fish4 = data1.loc[data1['Species'] == 'Parkki', ['Weight','Height'] ]

display( fish1.head( ) )
display( fish2.head( ) )
display( fish3.head( ) )
display( fish4.head( ) )
```

```python
#Step 10. 각 생선별 컬러를 지정하여 산점도(산포도)를 그립니다.
fig = plt.figure()

#문제 6.아래의 괄호 부분에 화면 분할하는 숫자를 적절하게 넣어 분할한 후
# 그래프를 예시화면과 같이 그리세요
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.scatter(fish1['Weight'], fish1['Height'] , color='red')
ax2.scatter(fish2['Weight'], fish2['Height'] , color='blue')
ax3.scatter(fish3['Weight'], fish3['Height'] , color='green')
ax4.scatter(fish4['Weight'], fish4['Height'] , color='deeppink')

#Step 11. 제목과 라벨명을 지정합니다.
ax1.set_title('Bream')
ax1.set_xlabel('Weight')
ax1.set_ylabel('Height')

ax2.set_title('Perch')
ax2.set_xlabel('Weight')
ax2.set_ylabel('Height')

ax3.set_title('Roach')
ax3.set_xlabel('Weight')
ax3.set_ylabel('Height')

ax4.set_title('Parkki')
ax4.set_xlabel('Weight')
ax4.set_ylabel('Height')

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.5, hspace=0.9)

plt.show()
```
