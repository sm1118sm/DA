```python
# Step 3. Species 값이 Bream 인 것만 추출하기
# 문제 4,5,6. if 문과 for 반복문과 리스트를 사용하여 위 data 변수 값에서
# Step 컬럼의 값 중 Etching , Cleaning , Deposition 데이터를 추출한 후
# 각 리스트에 할당한 후 각 공정별 데이터 건수를 출력하세요

list_etching = [ ]
list_cleaning = [ ]
list_deposition = [ ]

for a in data['Step'] :
  if a == 'Etching' :
    data_etching = data[data['Step'].isin(['Etching'])]
    list_etching.append(data_etching)
  elif a == 'Cleaning' :
    data_cleaning = data[data['Step'].isin(['Cleaning'])]
    list_cleaning.append(data_cleaning)
  elif a == 'Deposition' :
    data_deposition = data[data['Step'].isin(['Deposition'])]
    list_deposition.append(data_deposition)

len_etching = len(list_etching)
len_cleaning = len(list_cleaning)
len_deposition = len(list_deposition)

print('문제 4: Etching 공정 건수:' , len_etching)
print('문제 5: Cleaning 건수:' , len_cleaning)
print('문제 6: Deposition 건수:', len_deposition)
```
