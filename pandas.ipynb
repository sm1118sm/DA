```python
#주어진 파일 2개를 구글 코랩에서 불러온 후 각각 data1, data2 변수에 저장하세요. 그 후 파일 2개를 merge 작업을 수행하여 data3 변수에 저장하세요. 그 후 data3에서 판매번호, 제품명, 판매수량, 컬럼만 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
uploaded2 = files.upload()
file_name = list(uploaded.keys())[0]
file_name2 = list(uploaded2.keys())[0]
data1 = pd.read_csv(file_name, encoding = 'CP949')
data2 = pd.read_csv(file_name2, encoding = 'CP949')
data3 = pd.merge(data1, data2, left_on = '제품ID', right_on = 'ID', how = 'outer')
data3[['판매번호', '제품명', '판매수량']]
```

```python
#선수명에 '김'이 포함된 선수 중 홈런수가 20개 이상 50개 미만인 선수 조회하기

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_csv(file_name, encoding = 'CP949')
file1.loc[(file1['선수명'].str.contains('김')) & (file1['홈런'].between(20,50, inclusive ='left'))]
```

```python
#파일을 불러와서 메세지, 발생시간, 정지시간2, 작업구분 컬럼을 추출하되 정지시간2 값이 높은 순서대로 10건만 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_csv(file_name)
file1.sort_values('정지시간2', ascending = False)[['메세지', '발생시간', '정지시간2', '작업구분']].head(10)
```

```python
#파일을 활용하여 오류 메세지별 정지시간2 컬럼 값의 총합을 구한 후 정지시간2 값이 높은 순서부터 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_csv(file_name)
file1.groupby('메세지', as_index = False)['정지시간2'].sum().sort_values('정지시간2', ascending = False).head(10)
```

```python
#파일을 부러온 후 팀명을 기준으로 득점과 홈런의 sum과 mean 값을 구해서 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_excel(file_name)
data = pd.pivot_table(file1, index = '팀명', values = ['득점', '홈런'], aggfunc = ['sum', 'mean'])
data.sort_values(('sum','홈런'), ascending = False)
```

```python
#앞의 문제에서 mean의 홈런 개수가 20개 이상 50개 이하인 값만 추출한 후 mean 홈런 값이 큰 수부터 먼저 출력되도록 정렬하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_excel(file_name)
data = pd.pivot_table(file1, index = '팀명', values = ['득점', '홈런'], aggfunc = ['sum', 'mean'])
data2 = data.loc[(data['mean']['홈런'] >= 20) & (data['mean']['홈런'] <= 50)]
data2.sort_values(('mean', '홈런'), ascending = False)
```

```python
# 메세지,발생시간,정지시간2,작업구분,작업자 컬럼을 추출하되 작업자명이 홍길동이면서 정지시간2가 30이상 100미만인 값을 출력하세요.

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)
data['정지시간2'] = data['정지시간2'].fillna(0)
data.loc[
    (data['작업자'].isin(['홍길동'])) &
    (data['정지시간2'].between(30, 100, inclusive="left")),
    ['메세지', '발생시간', '정지시간2', '작업구분', '작업자']
]
```
