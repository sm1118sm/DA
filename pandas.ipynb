```python
#주어진 파일 2개를 구글 코랩에서 불러온 후 각각 data1, data2 변수에 저장하세요. 그 후 파일 2개를 merge 작업을 수행하여 data3 변수에 저장하세요. 그 후 data3에서 판매번호, 제품명, 판매수량, 컬럼만 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
uploaded2 = files.upload()
file_name = list(uploaded.keys())[0]
file_name2 = list(uploaded2.keys())[0]
data1 = pd.read_csv(file_name, encoding = 'CP949')
data2 = pd.read_csv(file_name2, encoding = 'CP949')
data3 = pd.merge(data1, data2, left_on = '제품ID', right_on = 'ID', how = 'outer')
data3[['판매번호', '제품명', '판매수량']]
```

```python
#파일을 불러와서 메세지, 발생시간, 정지시간2, 작업구분 컬럼을 추출하되 정지시간2 값이 높은 순서대로 10건만 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_csv(file_name)
file1.sort_values('정지시간2', ascending = False)[['메세지', '발생시간', '정지시간2', '작업구분']].head(10)
```

```python
#파일을 활용하여 오류 메세지별 정지시간2 컬럼 값의 총합을 구한 후 정지시간2 값이 높은 순서부터 출력하세요

import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file1 = pd.read_csv(file_name)
df_file1 = pd.DataFrame(file1)
df_file1['정지시간2'] = df_file1['정지시간2'].fillna(0)
df_file1.groupby('메세지', as_index = False)['정지시간2'].sum().sort_values('정지시간2', ascending = False).head(10)
```
